Missing são valores nulos, sem dados. Podemos tratar esses dados faltando no dataset, é uma questão muito importante que o cientista de dados tem que ficar muito atento

````python
df
````

````python
df.isna()

# o metodo isna vai retornar o dataset com o valor de cada celula sendo true ou false, se for true significa que aquele valor é nulo para aquela linha
````

Como vimos, no python true equivale a 1 e false a 0. se eu somar a coluna ou linha composta por true/false eu consigo saber quantos true existem.

o métodos .sum() soma todas as celulas de uma coluna
o .sum(1) soma todas as celulas de uma linha

````python
df.isna().sum()

# somando todos os NAN de uma coluna
# pega todos os valores de celulas NAN e conta
````

````python
df.isna().sum(1)

# verificando a quantidade de NAN por linha, ou seja, a soma de todas as celulas de cada linha
# sendo a soma de nans dos ativos para cada linha
````

# Transformacao Wide to Long

O melt é utilizado na biblioteca pandas para realizar uma transformacao que chamamos de wide to long. Ou seja, se tivermos um dataframe grande e largo podemos transformar num dataframe com poucas colunas, mas muitas linahas.

O que antes eram as colunas, agoram viram categorias em uma nova coluna, e colunas valores em uma nova coluna de valores


````python
cotacoes_ibov = pd.read_csv('cotacoes_ibov.csv', header = [0,1], index_col = 0)
````

````python
cotacoes_ibov.head()
````

````python
cotacoes_ibov.shape
````

fazendo o melt com o dataframe cotacoes

````python
cotacoes_ibov_longo = pd.melt(cotacoes_ibov, ignore_index = False)
````

````python
cotacoes_ibov_longo.head()
````

````python
cotacoes_ibov_longo.shape
````

Veja que com o melt, nos fomos de 546 colunas para apenas 3 colunas e de 247 linhas para 134.862 linhas, onde as colunas foram todas transformadas em linhas.

Assim tambem podemos aplicar filtros para diminuir a quantidade de linhas para cada categoria que foi criada a partir do melt

````python
cotacoes_ibov_longo_abev = cotacoes_ibov_longo[(cotacoes_ibov_longo['variable_0'] == 'Close') & (cotacoes_ibov_longo['variable_1'] == 'ABEV3.SA')]
````

````python
cotacoes_ibov_longo_abev.plot()
````

podemos alterar o nome das colunas

````python
cotacoes_ibov_longo.columns = ['Parametro', 'Ativo', 'Valor']
````
